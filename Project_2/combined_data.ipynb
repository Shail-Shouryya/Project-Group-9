```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
energy_data = "Resources/Project_2_energy_consumption_USA.csv"
energy_data = pd.read_csv(energy_data)
#energy_data.head()
investor_sentiment = "Resources/Project_2_investor_sentiment_USA.csv"
investor_sentiment = pd.read_csv(investor_sentiment)
investor_sentiment.head()
```

```python
energy_data = energy_data.rename(columns={"Value": "US Energy Consumption"})
energy_indexed = energy_data.set_index("Date")
#energy_indexed.head()
```

```python
investor_sentiment = investor_sentiment.rename(columns={"Bullish": "Bullish Sentiment", "Neutral": "Neutral Sentiment", "Bearish": "Bearish Sentiment"})
investor_condensed =investor_sentiment.drop(columns=["Total", "Bullish 8-Week Mov Avg", "Bullish Average", "Bullish Average + St. Dev", "Bullish Average - St. Dev", "S&P 500 Weekly High", "S&P 500 Weekly Low"])
investor_indexed = investor_condensed.set_index("Date")
investor_indexed.head()
```

```python
merged_data = pd.merge(energy_indexed, investor_indexed, on="Date")
merged_data.head()
```
